````python
import speech_recognition as sr
recognizer = sr.Recognizer()
with sr.Microphone() as source:
        print("Adjusting noise...")
        recognizer.adjust_for_ambient_noise(source, duration=1)
        print("Recording for 4 seconds...")
        recorded_audio = recognizer.listen(source, timeout=4)
        
        print("Done recording.")
````

````python
import os
from groq import Groq
client = Groq(
    api_key=api_key,
)

chat_completion = client.chat.completions.create(
    messages=[
                {
                    "role": "system",
                    "content": "Write the function in a Python code block with all necessary imports and no example usage.\n Note: I want to show the result in code",
                },
                {
                    "role": "user",
                    "content": 'write code to display time',
                },
            ],
    model="mixtral-8x7b-32768",
)

data = chat_completion.choices[0].message.content
````

````python
import re
def extract_code_from_result(result):
    pattern = r'```(?:\s*(?:py|python)\s*)?\n([\s\S]*?)```'
    match = re.search(pattern, result)
    if match:
        return match.group(1)
    return result  
data = extract_code_from_result(data)
data
````
